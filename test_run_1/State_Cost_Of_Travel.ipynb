```python
merge_1 = pd.merge(acc, att, how='left', on='merch_state', suffixes=("_acc", "_att"))
```

```python
merge_2 = pd.merge(rest, ret, how='left', on='merch_state', suffixes=("_rest", "_ret"))
```

```python
final = pd.merge(merge_1, merge_2, how='left', on='merch_state')
```

```python
final.head()
```

```python
final.info()
```

```python
trim = final[['merch_state', 'accommodation_cost', 'attraction_cost', 'breakfast_lunch_cost', 'dinner_cost', 'median_meal_cost', 
              'retail_cost']]
```

```python
trim.head()
```

```python
trim['cost_per_trip'] = (trim['accommodation_cost']*2)+ (trim['attraction_cost']*2)+ (trim['breakfast_lunch_cost']*4)+ (trim['dinner_cost']*2)+ (trim['retail_cost']*2)
```

```python
trim.head()
```

```python
trim = trim.loc[trim['merch_state'] !='XX']
```
